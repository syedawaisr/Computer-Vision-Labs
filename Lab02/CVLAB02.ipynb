```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
import cv2 as cv
img = cv.imread("/content/chess board.jpg")
# Convert the image to grayscale
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```python
img.shape
```

```python
# dt = cv.corneritems(img,2,3,0.04) # 2=(3,3)=block_size , 3=sobel, k=0.04
```

```python
dst = cv.cornerHarris(gray, blockSize=3, ksize=3, k=0.04)
# Threshold to get strong corners
threshold = 0.1* dst.max()
img[dst > threshold] = [0, 0, 255]  # Mark corners in red (BGR)
cv2_imshow(img)
```
